```python
#importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset= load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()

#Data cleanup
df['job_posted_date']= pd.to_datetime(df['job_posted_date'])
```

```python
df_salary= df[pd.notna(df['salary_year_avg'])].copy()

def projected_salary( salary):
    return print(f" your salary is {salary} your projected salary is: {salary * 1.03}" )


df_salary['salary_year_avg'].apply(projected_salary)

```

```python
import ast
def clean_up(skill):
    if pd.notna(skill):
        return ast.literal_eval(skill)


df['job_skills']= df['job_skills'].apply(clean_up)


df['job_skills'][0:5]
```

```python
def new_salary(row):
    if "Senior" in row["job_title_short"]:
        return 1.05* row['salary_year_avg']
    else:
        return 1.03*row['salary_year_avg']


df_salary['inflated_salary']= df_salary.apply(new_salary, axis= 1)

df_salary[['job_title_short', 'salary_year_avg', 'inflated_salary']]
```

```python
df['inflated_salary'] = new_salary(df_salary)

df_salary[['job_title_short', 'salary_year_avg', 'inflated_salary']]
```
