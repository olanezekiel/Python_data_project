```python
#importing libraries
import ast 
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset= load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()

#Data cleanup
df['job_posted_date']= pd.to_datetime(df['job_posted_date'])
df['job_skills']= df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_DA= df[df['job_title_short']== "Data Analyst"].copy()

```

```python
df_DA.head(5)
```

```python
df_DA['job_posted_month_no']= df_DA['job_posted_date'].dt.month

df_DA_exploded= df_DA.explode('job_skills')

df_DA_exploded
```

```python
DA_pivot= df_DA_exploded.pivot_table(index= 'job_posted_month_no', columns= 'job_skills', aggfunc= 'size', fill_value= 0)
```

```python
DA_pivot.loc['total']= DA_pivot.sum()

dA_pivot_sorted= DA_pivot[DA_pivot.loc['total'].sort_values(ascending=False).index].drop('total')
```

```python
dA_pivot_sorted
```

```python
dA_pivot_sorted.iloc[:, :6].plot(kind='line')
plt.title('Top 6 skills for Data Analyst by No')
plt.ylabel('Counts')
plt.xlabel('')
plt.show()
```
