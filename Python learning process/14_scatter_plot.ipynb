```python
#importing libraries
import ast 
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset= load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()

#Data cleanup
df['job_posted_date']= pd.to_datetime(df['job_posted_date'])
df['job_skills']= df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.head(3)
```

```python
df= df[df['job_title_short']== "Data Analyst"]

df_explode= df.explode('job_skills')
```

```python
skills_stat= df_explode.groupby('job_skills').agg(
    skills_count=('job_skills','count'),
    median_salary=('salary_year_avg','median')
)

skills_stat= skills_stat.sort_values(by='skills_count',ascending=False).head(10)
```

```python
skills_stat.plot(kind='scatter',x='skills_count', y='median_salary')
plt.xlabel('Job Skills Count')
plt.ylabel('Yearly Median Salary')
plt.title('Salary vs skills count for Data Analyst')
plt.tight_layout()

for i, txt in enumerate(skills_stat.index):
    plt.text(skills_stat['skills_count'].iloc[i],skills_stat['median_salary'].iloc[i],txt)
plt.show()
```
