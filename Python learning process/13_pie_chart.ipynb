```python
#importing libraries
import ast 
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset= load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()

#Data cleanup
df['job_posted_date']= pd.to_datetime(df['job_posted_date'])
df['job_skills']= df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', title='job_work_from_home',startangle=90, autopct= '%1.1f%%')
```

```python
df[['job_work_from_home','job_no_degree_mention','job_health_insurance']]
```

```python


fig, ax= plt.subplots(1,3)

job_dict= {'job_work_from_home':'Work from home',
      'job_no_degree_mention': 'No degree reguired',
      'job_health_insurance': 'health insurance'
      }





for i, (column, title)in enumerate(job_dict.items()):
    ax[i].pie(df[column].value_counts(),startangle=90, autopct= '%1.1f%%')
    ax[i].set_title(title)



plt.show()


```
